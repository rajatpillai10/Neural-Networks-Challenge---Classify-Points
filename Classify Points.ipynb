```python
import pandas as pd
import numpy as np
```

```python
X_Train = pd.read_csv("Logistic_X_Train.csv").values
X_Train.shape
```

```python
Y_Train = pd.read_csv("Logistic_Y_Train.csv").values
Y_Train.shape
```

```python
X_Test = pd.read_csv("Logistic_X_Test.csv").values
```

```python
X_Test.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X_Train[:,0],X_Train[:,1],c=Y_Train)
plt.show()
```

```python
from keras import models
from keras.layers import Dense
```

```python
model= models.Sequential()
model.add(Dense(20,activation='relu',input_shape=(2,)))
model.add(Dense(20,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
x_val=X_Train[:500]
x_train_new = X_Train[500:]
y_val=Y_Train[:500]
y_tain_new=Y_Train[500:]
```

```python
hist=model.fit(x_train_new,y_tain_new,epochs=40,batch_size=32,validation_data=(x_val,y_val))
```

```python
h=hist.history
h
```

```python
plt.plot(h['val_loss'],label="Validation Loss")
plt.plot(h['loss'],label="Training Loss")
plt.xlabel("EPOCHS")
plt.ylabel("LOSS")
plt.legend()
plt.show()
```

```python
plt.plot(h['val_accuracy'],label="Validation Accuracy")
plt.plot(h['accuracy'],label="Training Accuracy")
plt.xlabel("EPOCHS")
plt.ylabel("ACCURACY")
plt.legend()
plt.show()
```

```python
model.evaluate(X_Train,Y_Train)
```

```python
model.evaluate(X_Train,Y_Train)[1]
```

```python
model.evaluate(x_val,y_val)[1]
```

```python
pred = model.predict(X_Test)
prediction = np.zeros(pred.shape)
pred
```
